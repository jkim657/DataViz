```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math


# Data on Back Bay water quality titled "Water Quality Data" gathered by the U.S. 
# Fish and Wildlife Service.  The dataset does not have a license, however as a
# dataset created by U.S. government employees it is considered a U.S. Government Work,
# the dataset is considered to be within the public domain.

# The dataset was retrieved from the U.S. Data Catalog
# https://catalog.data.gov/dataset/water-quality-data-a3c08

# Each column of the dataset represents different attributes of the data such as
# the site it was taken from, air and water temperatures, pH, and the date the data
# was recorded.
# Each row contains data taken from a specific site at a specific date, denoted by
# Site_ID and Read_Date.
# The dataset is 117 kb, containing over 2000 items split between five different sites.


df = pd.read_csv("BKB_WaterQualityData_2020084.csv")

#split dataset by water bodies
df_bay = df[(df["Site_Id"]=="Bay")].copy()
df_a = df[(df["Site_Id"]=="A")].copy()
df_b = df[(df["Site_Id"]=="B")].copy()
df_c = df[(df["Site_Id"]=="C")].copy()

#data for bay area----------------------------------------------------------
bay_oxygen = df_bay["Dissolved Oxygen (mg/L)"].to_numpy()
bay_oxy_mask = np.isfinite(bay_oxygen)
bay_ph = df_bay["pH (standard units)"].to_numpy()
bay_ph_mask = np.isfinite(bay_ph)
bay_temp = df_bay["Water Temp (?C)"].to_numpy()
bay_temp_mask = np.isfinite(bay_temp)

bay_oxygen_masked = bay_oxygen[bay_oxy_mask]
bay_oxygen_masked = np.sort(bay_oxygen_masked)

bay_ph_masked = bay_ph[bay_ph_mask]
bay_ph_masked = np.sort(bay_ph_masked)

bay_temp_masked = bay_temp[bay_temp_mask]
bay_temp_masked = np.sort(bay_temp_masked)

#data for A-pool----------------------------------------------------------
a_oxygen = df_a["Dissolved Oxygen (mg/L)"].to_numpy()
a_oxy_mask = np.isfinite(a_oxygen)
a_ph = df_a["pH (standard units)"].to_numpy()
a_ph_mask = np.isfinite(a_ph)
a_temp = df_a["Water Temp (?C)"].to_numpy()
a_temp_mask = np.isfinite(a_temp)

a_oxygen_masked = a_oxygen[a_oxy_mask]
a_oxygen_masked = np.sort(a_oxygen_masked)

a_ph_masked = a_ph[a_ph_mask]
a_ph_masked = np.sort(a_ph_masked)

a_temp_masked = a_temp[a_temp_mask]
a_temp_masked = np.sort(a_temp_masked)

#data for B-pool----------------------------------------------------------
b_oxygen = df_b["Dissolved Oxygen (mg/L)"].to_numpy()
b_oxy_mask = np.isfinite(b_oxygen)
b_ph = df_b["pH (standard units)"].to_numpy()
b_ph_mask = np.isfinite(b_ph)
b_temp = df_b["Water Temp (?C)"].to_numpy()
b_temp_mask = np.isfinite(b_temp)
b_temp = df_b["Water Temp (?C)"].to_numpy()
b_temp_mask = np.isfinite(b_temp)

b_oxygen_masked = b_oxygen[b_oxy_mask]
b_oxygen_masked = np.sort(b_oxygen_masked)

b_ph_masked = b_ph[b_ph_mask]
b_ph_masked = np.sort(b_ph_masked)

b_temp_masked = b_temp[b_temp_mask]
b_temp_masked = np.sort(b_temp_masked)

#data for C-pool----------------------------------------------------------
c_oxygen = df_c["Dissolved Oxygen (mg/L)"].to_numpy()
c_oxy_mask = np.isfinite(c_oxygen)
c_ph = df_c["pH (standard units)"].to_numpy()
c_ph_mask = np.isfinite(c_ph)
c_temp = df_c["Water Temp (?C)"].to_numpy()
c_temp_mask = np.isfinite(c_temp)

c_oxygen_masked = c_oxygen[c_oxy_mask]
c_oxygen_masked = np.sort(c_oxygen_masked)

c_ph_masked = c_ph[c_ph_mask]
c_ph_masked = np.sort(c_ph_masked)

c_temp_masked = c_temp[c_temp_mask]
c_temp_masked = np.sort(c_temp_masked)


#indices for first and third quartiles
def q1_index (n):
    return math.ceil((n+1) * 0.25)

def q3_index (n):
    return math.ceil((n+1) * 0.75)


# bay_oxygen_boxes = [
#     {
#         'med': np.median(bay_oxygen_masked),
#         'q1': bay_oxygen_masked[ q1_index(bay_oxygen_masked.shape[0]) ],
#         'q3': bay_oxygen_masked[ q3_index(bay_oxygen_masked.shape[0]) ],
#         'whislo': bay_oxygen_masked[0],
#         'whishi': bay_oxygen_masked[bay_oxygen_masked.shape[0] - 1],
#         'mean': np.mean(bay_oxygen_masked),
#         'label': "Back Bay"
#     }
# ]

# a_oxygen_boxes = [
#     {
#         'med': np.median(a_oxygen_masked),
#         'q1': a_oxygen_masked[ q1_index(a_oxygen_masked.shape[0]) ],
#         'q3': a_oxygen_masked[ q3_index(a_oxygen_masked.shape[0]) ],
#         'whislo': a_oxygen_masked[0],
#         'whishi': a_oxygen_masked[a_oxygen_masked.shape[0] - 1],
#         'mean': np.mean(a_oxygen_masked),
#         'label': "A-Pool"
#     }
# ]


#box plots for dissolved oxygen levels
fig, ax = plt.subplots()

#dataframe dictionary for multiple boxplots
def box_dict(boxes):
    return {
        'med': boxes['meds'],
        'q1': boxes['q1s'],
        'q3': boxes['q3s'],
        'whislo': boxes['whislos'],
        'whishi': boxes['whishis'],
        'mean': boxes['means'],
        'label': boxes['labels']
    }

#dataframe containing data for multiple boxplots
boxes = pd.DataFrame({
    'meds': [np.median(bay_oxygen_masked), np.median(a_oxygen_masked), np.median(b_oxygen_masked), np.median(c_oxygen_masked)],
    'q1s': [bay_oxygen_masked[q1_index(bay_oxygen_masked.shape[0])], a_oxygen_masked[q1_index(a_oxygen_masked.shape[0])], b_oxygen_masked[q1_index(b_oxygen_masked.shape[0])], c_oxygen_masked[q1_index(c_oxygen_masked.shape[0])]],
    'q3s': [bay_oxygen_masked[q3_index(bay_oxygen_masked.shape[0])], a_oxygen_masked[q3_index(a_oxygen_masked.shape[0])], b_oxygen_masked[q3_index(b_oxygen_masked.shape[0])], c_oxygen_masked[q3_index(c_oxygen_masked.shape[0])]],
    'whislos': [bay_oxygen_masked[0], a_oxygen_masked[0], b_oxygen_masked[0], c_oxygen_masked[0]],
    'whishis': [bay_oxygen_masked[bay_oxygen_masked.shape[0] - 1], a_oxygen_masked[a_oxygen_masked.shape[0] - 1], b_oxygen_masked[b_oxygen_masked.shape[0] - 1], c_oxygen_masked[c_oxygen_masked.shape[0] - 1]],
    'means': [np.mean(bay_oxygen_masked), np.mean(a_oxygen_masked), np.mean(b_oxygen_masked), np.mean(c_oxygen_masked)],
    'labels': ["Back Bay", "A-Pool", "B-Pool", "C-Pool"]
})


plt.title("Dissolved Oxygen Levels in Back Bay")
plt.xlabel("Location")
plt.ylabel("Dissolved Oxygen (mg/L)")
#display each box plot in the dataframe boxes
ax.bxp([box_dict(boxes.iloc[i,:]) for i in range(len(boxes))], showmeans = True, showfliers = False)
plt.show()
plt.clf()


#Box plots for pH
fig, ax = plt.subplots()

#dataframe containing data for multiple boxplots
boxes = pd.DataFrame({
    'meds': [np.median(bay_ph_masked), np.median(a_ph_masked), np.median(b_ph_masked), np.median(c_ph_masked)],
    'q1s': [bay_ph_masked[q1_index(bay_ph_masked.shape[0])], a_ph_masked[q1_index(a_ph_masked.shape[0])], b_ph_masked[q1_index(b_ph_masked.shape[0])], c_ph_masked[q1_index(c_ph_masked.shape[0])]],
    'q3s': [bay_ph_masked[q3_index(bay_ph_masked.shape[0])], a_ph_masked[q3_index(a_ph_masked.shape[0])], b_ph_masked[q3_index(b_ph_masked.shape[0])], c_ph_masked[q3_index(c_ph_masked.shape[0])]],
    'whislos': [bay_ph_masked[0], a_ph_masked[0], b_ph_masked[0], c_ph_masked[0]],
    'whishis': [bay_ph_masked[bay_ph_masked.shape[0] - 1], a_ph_masked[a_ph_masked.shape[0] - 1], b_ph_masked[b_ph_masked.shape[0] - 1], c_ph_masked[c_ph_masked.shape[0] - 1]],
    'means': [np.mean(bay_ph_masked), np.mean(a_ph_masked), np.mean(b_ph_masked), np.mean(c_ph_masked)],
    'labels': ["Back Bay", "A-Pool", "B-Pool", "C-Pool"]
})


plt.title("pH of Back Bay")
plt.xlabel("Location")
plt.ylabel("pH (Standard Units)")
#display each box plot in the dataframe boxes
ax.bxp([box_dict(boxes.iloc[i,:]) for i in range(len(boxes))], showmeans = True, showfliers = False)
plt.show()
plt.clf()

#Box plots for water temperature
fig, ax = plt.subplots()

#dataframe containing data for multiple boxplots
boxes = pd.DataFrame({
    'meds': [np.median(bay_temp_masked), np.median(a_temp_masked), np.median(b_temp_masked), np.median(c_temp_masked)],
    'q1s': [bay_temp_masked[q1_index(bay_temp_masked.shape[0])], a_temp_masked[q1_index(a_temp_masked.shape[0])], b_temp_masked[q1_index(b_temp_masked.shape[0])], c_temp_masked[q1_index(c_temp_masked.shape[0])]],
    'q3s': [bay_temp_masked[q3_index(bay_temp_masked.shape[0])], a_temp_masked[q3_index(a_temp_masked.shape[0])], b_temp_masked[q3_index(b_temp_masked.shape[0])], c_temp_masked[q3_index(c_temp_masked.shape[0])]],
    'whislos': [bay_temp_masked[0], a_temp_masked[0], b_temp_masked[0], c_temp_masked[0]],
    'whishis': [bay_temp_masked[bay_temp_masked.shape[0] - 1], a_temp_masked[a_temp_masked.shape[0] - 1], b_temp_masked[b_temp_masked.shape[0] - 1], c_temp_masked[c_temp_masked.shape[0] - 1]],
    'means': [np.mean(bay_temp_masked), np.mean(a_temp_masked), np.mean(b_temp_masked), np.mean(c_temp_masked)],
    'labels': ["Back Bay", "A-Pool", "B-Pool", "C-Pool"]
})


plt.title("Water Temperature in Back Bay")
plt.xlabel("Location")
plt.ylabel("Water Temperature (CÂ°)")
plt.ylim([-5,100])
#display each box plot in the dataframe boxes
ax.bxp([box_dict(boxes.iloc[i,:]) for i in range(len(boxes))], showmeans = True, showfliers = False)
plt.show()
plt.clf()
```
