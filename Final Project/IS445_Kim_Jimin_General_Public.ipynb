```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


#Data on Back Bay water quality created by the U.S. Fish and Wildlife Service
#https://catalog.data.gov/dataset/water-quality-data-a3c08


df = pd.read_csv("BKB_WaterQualityData_2020084.csv")

#convert strings for read dates to datetime
df["Read_Date"]= pd.to_datetime(df["Read_Date"])

#split dataset by water bodies
df_bay = df[(df["Site_Id"]=="Bay")].copy()
df_a = df[(df["Site_Id"]=="A")].copy()
df_b = df[(df["Site_Id"]=="B")].copy()


#data for bay area----------------------------------------------------------
oxygen = df_bay["Dissolved Oxygen (mg/L)"].to_numpy()
oxy_mask = np.isfinite(oxygen)
ph = df_bay["pH (standard units)"].to_numpy()
ph_mask = np.isfinite(ph)
dates = df_bay["Read_Date"].to_numpy()

#Back Bay oxygen levels over time
dates_oxy_masked = dates[oxy_mask]
oxygen_masked = oxygen[oxy_mask]
sort_inds = dates_oxy_masked.argsort()
sort_inds = np.flip(sort_inds)
dates_oxy_masked = dates_oxy_masked[sort_inds[::-1]]
oxygen_masked = oxygen_masked[sort_inds[::-1]]

plt.figure(figsize=(12,8))
plt.title("Back Bay Dissolved Oxygen Levels Over Time")
plt.xlabel("Years")
plt.ylabel("Dissolved Oxygen (mg/L)")
plt.grid(True)
plt.plot(dates_oxy_masked, oxygen_masked)
plt.show()
plt.clf()

#Back Bay pH over time
dates_ph_masked = dates[ph_mask]
ph_masked = ph[ph_mask]
sort_inds = dates_ph_masked.argsort()
sort_inds = np.flip(sort_inds)
dates_ph_masked = dates_ph_masked[sort_inds[::-1]]
ph_masked = ph_masked[sort_inds[::-1]]

plt.figure(figsize=(12,8))
plt.title("Back Bay pH Over Time")
plt.xlabel("Years")
plt.ylabel("pH (Standard Units)")
plt.grid(True)
plt.ylim((0,10))
plt.yticks(np.arange(0, 10, step=1))
plt.plot(dates_ph_masked, ph_masked, color="#b08010")
plt.show()
plt.clf()



#data for A-Pool--------------------------------------------------
oxygen = df_a["Dissolved Oxygen (mg/L)"].to_numpy()
oxy_mask = np.isfinite(oxygen)
ph = df_a["pH (standard units)"].to_numpy()
ph_mask = np.isfinite(ph)
dates = df_a["Read_Date"].to_numpy()

#A-Pool oxygen levels over time
dates_oxy_masked = dates[oxy_mask]
oxygen_masked = oxygen[oxy_mask]
sort_inds = dates_oxy_masked.argsort()
sort_inds = np.flip(sort_inds)
dates_oxy_masked = dates_oxy_masked[sort_inds[::-1]]
oxygen_masked = oxygen_masked[sort_inds[::-1]]

plt.figure(figsize=(12,8))
plt.title("Back Bay A-Pool Dissolved Oxygen Levels Over Time")
plt.xlabel("Years")
plt.ylabel("Dissolved Oxygen (mg/L)")
plt.grid(True)
plt.plot(dates_oxy_masked, oxygen_masked)
plt.show()
plt.clf()


#A-Pool pH over time
dates_ph_masked = dates[ph_mask]
ph_masked = ph[ph_mask]
sort_inds = dates_ph_masked.argsort()
sort_inds = np.flip(sort_inds)
dates_ph_masked = dates_ph_masked[sort_inds[::-1]]
ph_masked = ph_masked[sort_inds[::-1]]

plt.figure(figsize=(12,8))
plt.title("Back Bay A-Pool pH Over Time")
plt.xlabel("Years")
plt.ylabel("pH (Standard Units)")
plt.grid(True)
plt.ylim((0,10))
plt.yticks(np.arange(0, 10, step=1))
plt.plot(dates_ph_masked, ph_masked, color="#b08010")
plt.show()
plt.clf()

#data for B-Pool--------------------------------------------------
oxygen = df_b["Dissolved Oxygen (mg/L)"].to_numpy()
oxy_mask = np.isfinite(oxygen)
ph = df_b["pH (standard units)"].to_numpy()
ph_mask = np.isfinite(ph)
dates = df_b["Read_Date"].to_numpy()

#B-Pool oxygen levels over time
dates_oxy_masked = dates[oxy_mask]
oxygen_masked = oxygen[oxy_mask]
sort_inds = dates_oxy_masked.argsort()
sort_inds = np.flip(sort_inds)
dates_oxy_masked = dates_oxy_masked[sort_inds[::-1]]
oxygen_masked = oxygen_masked[sort_inds[::-1]]

plt.figure(figsize=(12,8))
plt.title("Back Bay B-Pool Dissolved Oxygen Levels Over Time")
plt.xlabel("Years")
plt.ylabel("Dissolved Oxygen (mg/L)")
plt.grid(True)
plt.plot(dates_oxy_masked, oxygen_masked)
plt.show()
plt.clf()


#B-Pool pH over time
dates_ph_masked = dates[ph_mask]
ph_masked = ph[ph_mask]
sort_inds = dates_ph_masked.argsort()
sort_inds = np.flip(sort_inds)
dates_ph_masked = dates_ph_masked[sort_inds[::-1]]
ph_masked = ph_masked[sort_inds[::-1]]

plt.figure(figsize=(12,8))
plt.title("Back Bay B-Pool pH Over Time")
plt.xlabel("Years")
plt.ylabel("pH (Standard Units)")
plt.grid(True)
plt.ylim((0,10))
plt.yticks(np.arange(0, 10, step=1))
plt.plot(dates_ph_masked, ph_masked, color="#b08010")
plt.show()
plt.clf()
```
