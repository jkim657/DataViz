```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes as axes

df = pd.read_csv("USW00023185.csv")

temperatures = df["tmax"].to_numpy()

dates = pd.to_datetime(df["Date"])

dates = dates.to_numpy()


plt.title("Maximum Recorded Temperatures at Reno Airport, NV in 1893")
plt.ylabel("Maximum Recorded Temperature (FÂ°)")
plt.xlabel("Date")
plt.plot(dates, temperatures)
```
